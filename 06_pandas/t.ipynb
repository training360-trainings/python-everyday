```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. CSV fájl betöltése
# Helyettesítsd a 'your_file.csv' részt a saját fájlod nevével!
df = pd.read_csv("../files/eth.csv")

# Adatok első néhány sorának megtekintése
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d %I-%p")

# Ellenőrzés
print("Dátum oszlop típusa:", df["Date"].dtype)
df.head()
```

```python
numeric_cols = df.select_dtypes(include=["number"]).columns
print(f"Numerikus oszlopok: {list(numeric_cols)}")

# Újramintavételezés csak a numerikus oszlopokra
hourly_df = df[numeric_cols].resample("1h").mean()

hourly_df.head()
```
